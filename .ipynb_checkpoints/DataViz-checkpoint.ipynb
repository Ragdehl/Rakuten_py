## Image size analysis

The following code gets the size in MB of each picture:

```python
image_size_list = []
image_name_list = []

with os.scandir(images_path) as dir_contents:
    for entry in dir_contents:
        info = entry.stat()
        image_size_list.append(info.st_size)
        image_name_list.append(entry.name)
```

Compilethe size of each picture in a DataFrame:

```python
image_size_df = pd.DataFrame(image_size_list,columns=['image size bits'])
image_size_df

image_name_df = pd.DataFrame(image_name_list,columns=['image name'])
image_name_df

image_df = pd.concat([image_size_df,image_name_df],axis=1)
image_df.head()
```

Pictures distribution analysis:

```python
image_size_df.describe()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

image_size_df.plot.hist(y=['image size bits'], bins =40, rwidth = 0.8 , alpha=0.5);
```

It would be interesting to see if the size of the images varies depending on the type of product:

```python
import seaborn as sns
import matplotlib.pyplot as plt

X_train['y_train'] = y_train #on rajoute y_train dans X_train temporairement pour simplifier la tache

plt.figure(figsize=(20,8))
sns.boxplot(x='y_train',y='image size bits',data=X_train);

#Avec ce boxplot on peut voir que la taille des images varie en fonction du type de produit.
#Il pourrait donc être intéressant de garder cette variable pour categoriser.
```

At first glance we observe tha size distribution may depend on the category. 
To confirm or reject this hypothesis an ANOVA analysis (for qualitative) variables is launched:

**QUANTITATIVE & QUALITATIVE VARIABLES**

Analysis of variance (ANOVA). The objective of this test is to conclude on the influence of a categorical explanatory variable on the law of a continuous variable to be explained.

The reasoning we do with ANOVA is that if the qualitative variable (y_train: prdtypecode) does not affect the quantitative variable (image_size_bytes), the mean should be the same for the 27 modalities, i.e. μ1 = μ2 = ... = μ14μ1 = μ2 = ... = μ27.

H0: "μ1 = μ2 = ... = μ14μ1 = μ2 = ... = μ27"

(We reject H0 if the p-value is less than 5%: Reject H0 here means reject the hypothesis according to which the qualitative variable (y_train: prdtypecode) does not influence the quantitative variable (image_size_bytes))

```python
import statsmodels.api

X_train = X_train.rename(columns={'image size bits': 'image_size_bits'})

df = pd.concat([X_train,y_train],axis=1)

result = statsmodels.formula.api.ols('image_size_bits ~ prdtypecode', data = df).fit()
table = statsmodels.api.stats.anova_lm(result)
table
```

The p-value (PR (> F)) is less than 5%, then we reject the hypothesis according to which the qualitative variable (y_train = prdtypecode) does not affect the quantitative variable (image_size_bytes).

We can therefore consider that the size of an image varies according to the category.
The size of the images is therefore an important variable to take into account for a classification of data by category.

```python
# Visualisation des images
X_train['image name'] = 'image_' + X_train['imageid'].map(str) + '_product_' + X_train['productid'].map(str) + '.jpg'
plt.gray()
for category in list(set(y_train['prdtypecode'].tolist())):
    #print(category)
    f, axarr = plt.subplots(1,10,figsize=(15,15),subplot_kw = dict(xticks=[], yticks=[])) 
    
    pic_path_list = X_train[y_train['prdtypecode']==category]['image name'].head(10).tolist()
    i=0

    for pic in pic_path_list:
        picture = plt.imread(images_path + pic);
        #axarr[i].axis('off')
        axarr[i].set_ylabel(category)
        axarr[i].imshow(picture);
        i +=1
```
